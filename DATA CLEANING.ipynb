changing the column names to avoid white spaces

```python
weather_data.columns = ['Date', 'Summary', 'Precip_Type', 'Temperature',
       'Apparent_temperature', 'Humidity', 'Wind_speed',
       'Wind_Bearing', 'Visibility', 'Loud_Cover',
       'Pressure', 'Daily_Summary']
```

Since we have some nan values in Precip Type..we will be dropping that data

```python
weather_data[weather_data['Precip_Type'] == 'snow'].head()
```

```python
weather_data[weather_data['Precip_Type'] == 'rain'].head()
```

```python
weather_data.describe()
```

```python
weather_data.corr()
```

```python
weather_data['Daily_Summary'].nunique()
```

Too much values to categorise for the Daily Summary column

```python
weather_data =weather_data[weather_data['Precip_Type'].notnull()]
```

```python
weather_data.info()
```

```python
weather_data['Precip_Type'].isna().sum()
```

All rows with the null values have been dropped, unfortunately we could not fill the missing rows with median or mean values because it is a categorical data type not continuous

The next thing is to create new columns for whether teh Precipitation type is snow or rain

```python
weather_data['Precipitation_cat'] = [1 if x == 'rain' else 0 for x in weather_data['Precip_Type']]
```

```python
weather_data.info()
```

```python
weather_data.corr()
```

The next step is to drop the column loud cover since all the values are nan values and it won't be of much help in the EDA..

```python
del weather_data['Loud_Cover']
```

```python
weather_data.head()
```

```python
weather_data.corr()
```

We have got 7 continuous variables that can be used in predicting the target variable Apparent Temperature for our model

changing the date column from dataframe object to datetime

```python
weather_data['Date'] = weather_data['Date'].str.rstrip('+0200')
```

```python
weather_data.info()
```

```python
weather_data['Date'].dtype
```

```python
weather_data['Date'] = pd.to_datetime(weather_data['Date'], infer_datetime_format = True, utc = True)
```

```python
weather_data.info()
```

```python
weather_data.head()
```

#### Next step is to drop the duplicated data from the dataframe

```python
weather_data.duplicated().sum()
```

```python
weather_data.drop_duplicates(inplace=True)
```

```python
weather_data.duplicated().sum()
```

```python
cleaned_data = weather_data
```

```python
cleaned_data.tail()
```

```python
cleaned_data.to_csv('cleaned_data', encoding = 'utf-8', index = False)
```
