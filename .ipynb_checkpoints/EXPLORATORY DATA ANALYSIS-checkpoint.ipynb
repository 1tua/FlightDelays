### 1.Which airports have the highest and lowest average delay minutes amongst LAX(LOS ANGELES INTERNATIONAL), MIA(MIAMI INTERNATIONAL), ORD(O'HARE INETERNATIONAL(CHICAGO)), DFW(DALLAS/FORT WORTH INTERNATIONAL), JFK( JOHN F KENNEDY INTERNATIONAL (NEW YORK))?

### 2.Which months have the highest and lowest average delay minutes amongst the months of travel?

```python
plt.figure(figsize =(16,6))
sns.boxplot(x="Month", y="Delay_min", data=data, showfliers = False);
plt.title('Categorical plots of delay minutes across all the 10 months of travel recorded');
```

We can glean from the plot above that the highest average in terms of delay minutes occurs in the month of June and lowest occurs in the month of either September or February. To find out for certain..we can call the describe method on the dataframes

```python
data[data['Month'] == 2].describe()
```

```python
data[data['Month'] == 9].describe()
```

### 3.Which days have the highest and lowest average delay minutes amongst the days of travel?

```python
plt.figure(figsize=(16, 6))
fig.set_size_inches(11.7, 8.27)
sns.boxplot(x="DayofMonth", y="Delay_min" ,data=data, showfliers = False);
plt.title('Categorical plots of delay minutes across all the 28-31 days/dates of travel recorded');
```

There is a trend in the plot above as we see that the averages for each day of the month lie around zero at the beginning of the month and at the end of the month, the averages go down with less delays aas average but there are also strong outliers at both end of the spectrum, most especially on the 6th day of the month

### 4.Which airports have the highest and lowest average windspeed(in terms of how terrible the weather tends to be)?

We can definitely confirm from the plot above that the Dallas Forth Worth International, Texas has the worst conditions for wind speed and Los Angeles International airport has the best conditions in terms of the wind speed.
This might give us sa littele bit of information on which of the airports' services might get disrupted by weather

### 5.Which hours of the day have the highest and lowest average delay minutes across all the international airports ?

```python
plt.figure(figsize =(16,6))
sns.boxplot(x="Hour", y="Delay_min", data=data, showfliers = False);
plt.title('Categorical plots of Windspeed across all the airports');
```

## DATA DISTRIBUTION CHECKS FOR ALL THE ASSUMPTIONS OF LINEAR REGRESSION MODELLING

### MULTI-COLLINEARITY CHECK

```python
data.corr() *100# note that the target variable is delay minutes
```

The delay minutes correlates highly positively with departure delay and arrival delay

The next step is to check for Linearity amongst the independent and dependent variables

```python
fig = plt.figure(figsize = (20,10))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)


sns.scatterplot(x = data['Delay_min'],y = data['Temp'] ,color = 'red', ax = ax1)
sns.scatterplot(x= data['Delay_min'], y = data['Pressure'], color = 'blue', ax = ax2)
sns.scatterplot(x = data['Delay_min'], y = data['ArrDelay'], color = 'green', ax = ax3)
sns.scatterplot(x= data['Delay_min'], y = data['Wind'], color = 'yellow', ax = ax4);
```

Arrival Dealy is the only independent variable that shows linearity between the target variable and the independent variables>

## Checking for the normality of the distribution of the variables in the dataset

Plotting the distributions of the variables

```python
fig = plt.figure(figsize = (20,10))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

sns.distplot(data['Delay_min'], color = 'red', ax = ax1)
sns.distplot(data['Pressure'], color = 'blue', ax = ax2)
sns.distplot(data['ArrDelay'], color = 'green', ax = ax3)
sns.distplot(data['Wind'], color = 'yellow', ax = ax4);
```

It is evident from the plots above that none of the plots are normally distributed especially the wind distribution plot which looks more categorical than continuous

```python
fig = plt.figure(figsize = (20,10))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

sns.distplot(data['Temp'], color = 'red', ax = ax1)
sns.distplot(data['Hour'], color = 'blue', ax = ax2)
sns.distplot(data['Humidity'], color = 'green', ax = ax3)
sns.distplot(data['AirTime'], color = 'yellow', ax = ax4);
```

Temperature independent variable is the closest to a normal distribution for all the plots. We can definitley drop the variable arrival delay from the modelling because of auto correlation as the arrival delay and departure dalay are somewhat correlated

We can come to the conclusion that there is multicollinearity between variables to be used in the model, there is linearity between the ArrDelay and Delay_minutes but we cannot use the variable because of Autocorrelation. and none of the plots/varaibles satisfy the normality tests
