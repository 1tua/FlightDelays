#### RESIDUAL Q-Q PLOTS AND HOMOSCEDASTICITY TESTS

Running a simple regression model to check whether Delay minute is a good predictor of Arrival Delay

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# build the formula 
f = 'ArrDelay~Delay_min'
# create a fitted model in one line
model = smf.ols(formula=f, data=data).fit()
```

```python
model.summary()
```

This shows that the predictor variable Delay minute is a strong predictor of Arrival Delay.The R-squared value is 0.818 i.e. 81.8% of the variance in the target variable can be explained by Delay minutes. 

```python
X_new = pd.DataFrame({'Delay_min': [data.Delay_min.min(), data.Delay_min.max()]})
print(X_new.head())

# make predictions for those x values and store them
preds = model.predict(X_new)
print (preds)

# first, plot the observed data and the least squares line
data.plot(kind='scatter', x='Delay_min', y='ArrDelay')
plt.plot(X_new, preds, color='red', linewidth=2)
plt.show()
```

```python
#visulaizing the error term for variance and heteroscedasticity
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Delay_min", fig=fig)
plt.show()
```

This residual plot above is going by the homoscedasticity assumption for the predictor variable Delay minute

```python
#checking for the q-q plot
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

We can see from the above plot its does not check the linearity assumption as well based on the Q-Q plot
